# Descente de gradient dynamique
Une fonction sympa pour représenter graphiquement et en temps réel l'optimisation d'une fonction de coût (par exemple avec la méthode de descente de gradient). Le graphique s'actualise à chaque nouvelle itération.

```python
# Imports divers
import pylab as pl
from IPython import display
import matplotlib.pyplot as plt
import time 
plt.style.use('ggplot')

# Données simulées
x    = [1,2,3,4,5]
cost = [4,2.5,1.8,1.5,1.3]

for i in range(len(x)+1):
    pl.clf()
    plt.title('Descente de gradient')
    plt.xlabel('# Itérations')
    plt.ylabel('Coût ')
    plt.plot(x[:i], cost[:i], marker='o')
    plt.xlim((0, len(x)+1))
    plt.ylim((0, len(x)+1))
    display.display(pl.gcf())
    display.clear_output(wait=True)
    time.sleep(1.0)
```
